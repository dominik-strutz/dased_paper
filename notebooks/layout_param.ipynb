```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.style as style

style.use("https://raw.githubusercontent.com/dominik-strutz/dotfiles/main/mystyle.mplstyle")
```

```python
from dased.layout import DASLayout
```

```python
from dased.layout import DASLayout

knots = np.array(
    [[-200.0, -400.0], [-200.0, 1000.0], [1000.0, 1000.0], [2200.0, 2200.0]]
)

das_linear = DASLayout(
    knots,spacing=100, k=1)
das_cubic = DASLayout(
    knots,spacing=100, k=3)

display(das_cubic)
```

```python
das_cubic.get_gdf()
```

```python
# Define layout configurations
layouts = [das_linear, das_cubic]
titles = ["Linear DAS Layout", "Cubic DAS Layout"]
labels = ["channel", "channel"]
knot_labels = ["knot", "knot"]

fig, axs = plt.subplots(1, len(layouts), figsize=(len(layouts) * 3, 3))
fig.set_facecolor("white")

# Plot each layout
for i, (layout, title, label) in enumerate(zip(layouts, titles, labels)):
    # Plot the layout
    layout.plot(
        ax=axs[i],
        show_knots=True,
        label=label,
        color="blue",
        knot_kwargs={"label": knot_labels[i], "color": "black", "zorder": -1},
    )

    # Configure the axes
    axs[i].set_title(title)
    axs[i].set_xlabel("")
    if i == 0:
        axs[i].set_ylabel("")
    axs[i].set_aspect("equal", adjustable="box")
    axs[i].set_xticks([])
    axs[i].set_yticks([])

    axs[i].grid(True, alpha=0.3)

# Make all axes have the same limits for better comparison
xlim = [
    min([ax.get_xlim()[0] for ax in axs]) - 200,
    max([ax.get_xlim()[1] for ax in axs]) + 200,
]
ylim = [
    min([ax.get_ylim()[0] for ax in axs]) - 200,
    max([ax.get_ylim()[1] for ax in axs]) + 200,
]

for ax in axs:
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

# Add a single legend for all plots at the figure level
handles, labels = axs[0].get_legend_handles_labels()  # Get handles from first plot
for ax in axs[1:]:
    h, l = ax.get_legend_handles_labels()
    for handle, label in zip(h, l):
        if label not in labels:  # Avoid duplicates
            handles.append(handle)
            labels.append(label)

# Place legend at the top of the figure
fig.legend(
    handles,
    labels,
    loc="upper center",
    bbox_to_anchor=(0.5, 0.15),
    ncol=2,
    fontsize=8,
    handlelength=2.5,
)

fig.savefig('figures/layout_parametrisation.pdf', dpi=300)
fig.savefig('figures/layout_parametrisation.png', dpi=300)
plt.show()
```
